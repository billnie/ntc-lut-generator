```python
def voltsAt(idx):
    return selection[idx]

def tempAt(idx):
    return tempSelection[idx];

def binarySearchTemp(temp, startIdx, length):
    #print temp, startIdx, length
    if(1 == length):
        return startIdx;
    i = startIdx + length/2;
    if(temp > tempAt(i)):
        return binarySearchTemp(temp, i, length/2);
    elif temp < tempAt(i):
        return binarySearchTemp(temp, startIdx, length/2);
    else:
        return i

def binarySearchVolts(voltage, startIdx, length):
    #print temp, startIdx, length
    if(1 == length):
        return startIdx;
    i = startIdx + length/2;
    if(voltage > voltsAt(i)):
        return binarySearchVolts(voltage, i, length/2);
    elif voltage < voltsAt(i):
        return binarySearchVolts(voltage, startIdx, length/2);
    else:
        return i

def interpolate(val, rangeStart, rangeEnd, valStart, valEnd):
    return rangeStart + (rangeEnd - rangeStart) * (val - valStart) / (valEnd - valStart)
    
def interpolateVoltage(temp, i):
    if tempAt(i) > temp:
        return interpolate(temp, voltsAt(i-1), voltsAt(i), tempAt(i-1), tempAt(i))
    elif tempAt(i) < temp:
        return interpolate(temp, voltsAt(i), voltsAt(i+1), tempAt(i), tempAt(i+1))
    else:
        return voltsAt(i)

def interpolateTemperature(voltage, i):
    if voltsAt(i) > voltage:
        return interpolate(voltage, tempAt(i-1), tempAt(i), voltsAt(i-1), voltsAt(i))
    elif voltsAt(i) < voltage:
        return interpolate(voltage, tempAt(i), tempAt(i+1), voltsAt(i), voltsAt(i+1))
    else:
        return tempAt(i)
        
interpolated = np.zeros(1134)
for i in range(0, len(interpolated)):
    interpolated[i] = interpolateVoltage(i, binarySearchTemp(i,0,len(selection)))
    
error=millivolts[:1134]-interpolated

figure()
title("Voltage error when using 65 point look-up table")
xlabel("Temperature, C")
ylabel("Voltage, mV")
plot(abs(error[:]))
#for i in selectedIndices:
#    if(i < 200):
#        axvline(i, 0, 100, linewidth=1, color='red')
figure()
maxErrorC = max(abs(error/ratio[:1134]))
maxError = max(abs(error))
print "maxError:", maxError,"mV", maxErrorC, "C"

hist(error, bins=40)
deviation = np.std(error)
axvline(np.mean(error) - deviation, 0, 100, linewidth=1, color='red')
axvline(np.mean(error) + deviation, 0, 100, linewidth=1, color='red')


interpolatedTemp = np.zeros(len(temperatures))
for i in range(0, len(temperatures)):
    voltage = 5000000 / 2**14 * i / 101
    voltage = voltage/1000.
    interpolatedTemp[i] = interpolateTemperature(voltage, binarySearchVolts(voltage,0,len(selection)))

tempError = (temperatures - interpolatedTemp)[:14000]  
figure()
title("Temperature error when using 65 point look-up table")
xlabel("ADC value, LSB (Vref = 5V)")
ylabel("Temperature, C")

plot(tempError)

figure()
hist(tempError, bins=40)
deviation = np.std(tempError)
axvline(np.mean(tempError) - deviation, 0, 100, linewidth=1, color='red')
axvline(np.mean(tempError) + deviation, 0, 100, linewidth=1, color='red')

```
